# Importing python libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading dataset

```python
dataset=pd.read_csv(r"C:\Users\advai\Downloads\assignments\Salary_dataset.csv")
```

```python
dataset
```

```python
dataset = dataset.drop(columns=['Unnamed: 0'])
```

```python
dataset
```

# Checking null values

```python
dataset.isnull().sum()
```

# Slicing dataset in x and y

```python
x=dataset[["YearsExperience"]]
y=dataset[["Salary"]]
```

```python
plt.scatter(x,y)
plt.xlabel('YearsExperience', fontsize=20)
plt.ylabel('Salary', fontsize=20)
plt.show()
```

#  Spliting data in Train and test 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=11)
```

# Model Fitting 

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# Viewing the b1 and b0 value
weight = regressor.coef_
print("Weight of the model={}".format(weight))

bais = regressor.intercept_
print("Intercept of the model={}".format(bais))
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
import pandas as pd

results = pd.DataFrame({
    'YearsExperience': x_test.iloc[:, 0].values,
    'Actual_Salary': y_test.iloc[:, 0].values,
    'Predicted_Salary': y_pred.flatten()
})

print(results)
```

# Checking Accuracy_score

```python
from sklearn.metrics import r2_score
ab=r2_score(y_test,y_pred)*100
ab
```
